```python
# A) Contracts

pragma solidity ^0.8.0;

contract C {
    uint private data;
    uint public info;

    constructor() {
        info = 10;
    }

    function increment(uint a) private pure returns(uint) {
        return a + 1;
    }

    function updateData(uint a) public {
        data = a;
    }

    function getData() public view virtual returns(uint) {
        return data;
    }

    function compute(uint a, uint b) internal pure returns (uint) {
        return a + b;
    }
}

contract E is C {
    uint private result;
    C private c;

    constructor() {
        c = new C();
    }

    function getComputedResult() public {
        result = compute(3, 5);
    }

    function getResult() public view returns(uint) {
        return result;
    }

    function getData() public view override returns(uint) {
        return c.info();
    }
}

# B)	Inheritance:

pragma solidity >=0.8.0 <0.9.0;

// Defining contract
contract Parent {
    uint internal sum;

    // Defining external function to set value of internal state variable sum
    function setValue() external {
        uint a = 20;
        uint b = 20;
        sum = a + b;
    }
}

// Defining child contract
contract Child is Parent {
    // Defining external function to return value of internal state variable sum
    function getValue() external view returns(uint) {
        return sum;
    }
}

// Defining calling contract
contract Caller {
    Child cc;

    // Creating child contract object
    constructor() {
        cc = new Child();
    }

    // Defining function to call setValue and getValue functions
    function testInheritance() public {
        cc.setValue();
    }

    function result() public view returns(uint) {
        return cc.getValue();
    }
}


# C)	Constructors:
pragma solidity ^0.8.0;

contract Base {
    uint data;

    constructor(uint _data) {
        data = _data;
    }

    function getresult() public view returns (uint) {
        return data;
    }
}

contract Derived is Base {
    constructor(uint _info) Base(_info * _info) {}
}

# Indirect Initialization of Base Constructor

pragma solidity ^0.8.0;

contract Base {
    uint data;

    constructor(uint _data) {
        data = _data;
    }

    function getresult() public view returns(uint) {
        return data;
    }
}

contract Derived is Base {
    constructor(uint _info) Base(_info * _info) {}
}


# D)	Abstract Contracts:

pragma solidity ^0.8.0;

interface ICalculator {
    function getResult() external pure returns (uint);
}

contract Test is ICalculator {
    constructor() {}

    function getResult() external pure returns (uint) {
        uint a = 5;
        uint b = 2;
        uint result = a + b;
        return result;
    }
}


# E)	Interfaces:

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

interface Calculator {
    function getResult() external view returns(uint);
}

contract Test is Calculator {
    constructor() {}

    function getResult() external pure override returns(uint) {
        uint a = 5;
        uint b = 2;
        uint result = a + b;
        return result;
    }
}


# F)	Libraries:

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

library Search {
    function indexOf(uint[] storage self, uint value) public view returns (uint) {
        for (uint i = 0; i < self.length; i++) {
            if (self[i] == value) {
                return i;
            }
        }
        return type(uint).max; // Use type(uint).max instead of uint(-1)
    }
}

contract Test {
    uint[] data;
    uint value;
    uint index;

    constructor() {
        data.push(6);
        data.push(7);
        data.push(8);
        data.push(9);
        data.push(10);
    }

    function isValuePresent() external {
        value = 9;
        index = Search.indexOf(data, value);
    }

    function getresult() public view returns(uint) {
        return index;
    }
}


# G)	Assembly:

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

library Sum {
    function sumUsingInlineAssembly(uint[] memory _data) public pure returns (uint o_sum) {
        for (uint i = 0; i < _data.length; ++i) {
            assembly {
                o_sum := add(o_sum, mload(add(add(_data, 0x20), mul(i, 0x20))))
            }
        }
    }
}

contract Test {
    uint[] data;

    constructor() {
        data.push(1);
        data.push(2);
        data.push(3);
        data.push(4);
        data.push(5);
    }

    function sum() external view returns(uint) {
        return Sum.sumUsingInlineAssembly(data);
    }
}


# H)	Events:

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract EventExample {
    uint256 public value = 0;

    event Increment(address owner);

    function getValue(uint _a, uint _b) public {
        emit Increment(msg.sender);
        value = _a + _b;
    }
}


# I)	Error Handling:

# Solidity program to demonstrate require statement.

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract RequireStatement {
    function checkInput(uint8 _input) public pure returns(string memory) {
        require(_input <= 255, "invalid uint8");
        return "Input is Uint8";
    }

    function odd(uint _input) public pure returns(bool) {
        require(_input % 2 != 0, "Input is not odd");
        return true;
    }
}


# Solidity program to demonstrate assert statement.

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract AssertStatement {
    bool public result;

    function checkOverflow(uint8 _num1, uint8 _num2) public {
        uint8 sum = _num1 + _num2;
        assert(sum <= 255);
        result = true;
    }

    function getResult() public view returns(string memory) {
        return result ? "No Overflow" : "Overflow exists";
    }
}

# Solidity program to demonstrate revert statement.

// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract RevertStatement {
    function checkOverflow(uint _num1, uint _num2) public view returns (string memory, uint) {
        uint sum = _num1 + _num2;
        if (sum < 0 || sum > 255) {
            revert("Overflow Exist");
        } else {
            return ("No Overflow", sum);
        }
    }
}
```
